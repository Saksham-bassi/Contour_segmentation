```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path="C:/Users/User/OneDrive/Desktop/Untitled Folder/Screenshot 2024-07-11 141506.png"
image=cv2.imread(image_path)
plt.imshow(image)
print(image.shape)
```

```python
gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(gray_image,cmap="gray")
gray_image.max()
```

```python
_, binary_image = cv2.threshold(gray_image, 115, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image,cmap="gray")
```

```python
contours, _ = cv2.findContours(binary_image, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
len(contours)
```

```python
contour_image = np.zeros_like(binary_image)
```

```python
contours = sorted(contours, key=cv2.contourArea, reverse=True)
```

```python
mask = np.zeros_like(image)
# Draw the Second largest contour (rectangle)
plt.imshow(cv2.drawContours(mask, [contours[1]], -1, (255),thickness=cv2.FILLED),cmap="gray")
```

```python
gray_mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
```

```python
result = cv2.bitwise_and(image, image, mask=gray_mask)
result_rgb = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(result_rgb)
```
